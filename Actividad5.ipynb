1. Utilizando la herramienta Júpiter Notebook programa lo siguiente:

*   Escribe una lista que contenga un mínimo de cinco elementos.
*   Agrega un nuevo elemento en el índice 2.
*   Agrega un elemento al final de la lista.
*   Escribe una cadena de caracteres que contenga tres palabras.
*   Cuenta los caracteres de la cadena generada.
*   Guarda la cadena resultante toda en mayúsculas en una nueva variable.
*   Escribe un diccionario que contenga al menos cuatro elementos y cuyos índices coincidan con las primeras letras del alfabeto.
*   Agrega un nuevo elemento al diccionario.
*   Elimina el elemento del índice “c”.
*   Cuenta los elementos que componen el diccionario después de completadas las operaciones anteriores.

```python
# Definición de lista
Boybands = ["Seventeen", "BTS", "Big Time Rush", "StrayKids", "One Direction"]
print(Boybands)
print("-"*40)

# Agrega un nuevo elemento en el índice 2.
Boybands.insert(2, "One OK Rock")
print(Boybands)
print("-"*40)

# Agrega un elemento al final de la lista.
Boybands.append("Morat")
# Imprimir la lista completa
print(Boybands)
print("-"*40)

#Escribe una cadena de caracteres que contenga tres palabras.
Boybands.append("Python es intuitivo")
#Cuenta los caracteres de la cadena generada.
cadena = Boybands[-1]
cantidad = len(cadena)
print(f"La cadena es: '{cadena}'")
print(f"Tiene un total de {cantidad} caracteres.")
print("-"*40)

# Guarda la cadena resultante toda en mayúsculas en una nueva variable.
mayusculas = Boybands[-1].upper()
print(f"Variable: {Boybands[-1]}")
print(f"Nueva variable: {mayusculas}")
print("-"*40)

# Escribe un diccionario que contenga al menos cuatro elementos y cuyos índices coincidan con las primeras letras del alfabeto.
diccionario_alfabeto = {
    "a": "Avion",
    "b": "Burro",
    "c": "Casa",
    "d": "Dado"}
print(diccionario_alfabeto)
print("-"*40)

# Agrega un nuevo elemento al diccionario.
diccionario_alfabeto["e"] = "Elefante"
print(diccionario_alfabeto)
print("-"*40)

# Elimina el elemento del índice “c”.
del diccionario_alfabeto["c"]
print(diccionario_alfabeto)
print("-"*40)

# Cuenta los elementos que componen el diccionario después de completadas las operaciones anteriores.
total_elementos = len(diccionario_alfabeto)
print(f"El diccionario contiene {total_elementos} elementos.")
```

Con lo aprendido en clase, elije uno de los siguientes problemas e implementa un programa en Python con la solución:



*   Una sala de cine cobra diferentes precios de entrada según la edad de la persona. Si una persona es menor de 3 años, la entrada es gratuita; si tienen entre 3 y 12, el boleto es de $10; y si son mayores de 12 años, el boleto cuesta $ 15. Escribe un bucle en el que se solicite a los usuarios su edad y despliegue el costo de la entrada.
*   Crea un diccionario que contenga la información de una canción (artista, álbum, duración y género) y despliega el contenido de cada elemento (keys, values y tuplas)





```python
print("--- LYRISSE'S CINEMA ---")
print("(Escribe 'pagar' para terminar la compra)\n")

while True:
    entrada = input("Por favor, ingresa tu edad: ")

    # Condición para romper el bucle
    if entrada.lower() == 'pagar':
        print("Boletos pagados")
        print("¡Disfruta la película!")
        break

    # Convertimos la entrada a número entero
    edad = int(entrada)

    # Lógica de precios según la edad
    if edad < 3:
        precio = 0
    elif 3 <= edad <= 12:
        precio = 10
    else:
        precio = 15

    # Desplegar el costo
    if precio == 0:
        print("Tu entrada es GRATUITA.\n")
    else:
        print(f"El costo de tu boleto es de: ${precio}.\n")
```
